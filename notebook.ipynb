```python
# Create a list of dataframes
dfs = [aotizhongxin_df, changping_df, dingling_df, dongsi_df, guanyuan_df, gucheng_df,
       huairou_df, nongzhanguan_df, shunyi_df, tiantan_df, wanliu_df, wanshouxigong_df]

# Create a list of average PM2.5 values for each dataframe
pm25_avg = [df['PM2.5'].mean() for df in dfs]
pm10_avg = [df['PM10'].mean() for df in dfs]

# Add the average PM2.5 values to each dataframe
for i, df in enumerate(dfs):
    df['average_PM2.5'] = pm25_avg[i]
    df['average_PM10'] = pm10_avg[i]

# Concatenate the dataframes vertically
result_df = pd.concat(dfs, ignore_index=True)

result_df
```

```python
# Questions 01:
# What is the average PM2.5 level across all districts in China?

pm25_avg_df = result_df.groupby('station')['PM2.5'].mean().reset_index().rename(columns={'PM2.5': 'PM2.5 Avg'})
pm25_avg_df
```

```python
# Then, we visualize the answer of 1st question using bar chart
pm25_avg_df = pm25_avg_df.sort_values('PM2.5 Avg', ascending = True)
plt.barh(pm25_avg_df['station'], pm25_avg_df['PM2.5 Avg'])
plt.xlabel('PM2.5 Average')
plt.ylabel('District')
plt.title('Average PM2.5 by District')
plt.xticks()
plt.show()
```

```python
# Question 02:
# How does the average PM10 level vary across different in a specific district?

pm10_avg_df = result_df.groupby('station')['PM10'].mean().reset_index().rename(columns={'PM10': 'PM10 Avg'})
pm10_avg_df
```

```python
# Then, we visualize the answer of 2nd question using bar chart
pm10_avg_df = pm10_avg_df.sort_values('PM10 Avg', ascending = True)
plt.barh(pm10_avg_df['station'], pm10_avg_df['PM10 Avg'])
plt.xlabel('PM10 Average')
plt.ylabel('District')
plt.title('Average PM10 by District')
plt.xticks()
plt.show()
```

```python
# Question 03:
#  How do wind direction affect average PM2.5 levels?

wd_pm25_df = result_df.groupby('wd')['PM2.5'].mean().reset_index().rename(columns={'PM2.5': 'PM2.5 Avg'})
wd_pm25_df
```

```python
# Sort each district by PM10 values in descending order
sorted_districts = district_means.sort_values(by='PM10', ascending=False)
most_polluted_districts_in_2017 = sorted_districts.index[:3]

print("These are the top 3 most recent pollution events in 2017:")
for district in most_polluted_districts_in_2017:
    pm10_value = district_means.loc[district, 'PM10']
    rounded_pm10_value = round(pm10_value, 2)
    print('- ' + district + ' with PM10 values equal to ' + str(rounded_pm10_value))
```
